```python
### Working Version of Class ###
## Ahmet Faruk Kavak
## 2016703063

## Q1
class polynomial(object):
    def __init__(self, c, v):
        self.coeff = c
        self.v = v
    def __repr__(self):
        D = len(self.coeff)
        # below is needed for multiplication method
        ranGe=[i for i in range(len(self.coeff)) if self.coeff[i] != 0]
        for i in ranGe:
            if i<D-1:
                print(self.coeff[i], end='')
                print('{}^{} '.format(self.v, D-i-1), end='')
            else:
                print(self.coeff[i])
        
        return str(self.coeff)
    def __add__(self,b):
        """Computes a+b and returns the result"""
        D = len(self.coeff)
        D_ = len(b.coeff)
        if(D>D_) :
            for n in range(D-D_):
                b.coeff.insert(0,0)
        coeff = []
        for i in range(D):
            coeff.append(self.coeff[i] + b.coeff[i])
        
        return polynomial(coeff, self.v)
## Q2
    def __sub__(self,b):
        """Computes a-b and returns the result"""
        D = len(self.coeff)
        D_ = len(b.coeff)
        if(D>D_) :
            for n in range(D-D_):
                b.coeff.insert(0,0)
        coeff = []
        for i in range(D):
            coeff.append(self.coeff[i] - b.coeff[i])
        return polynomial(coeff,self.v)
## Q3
    def __mul__(self,b):
        """Computes a*b and returns the result"""
        D = len(self.coeff)
        D_ = len(b.coeff)
        if(D>D_) :
            for n in range(D-D_):
                b.coeff.insert(0,0)
        c1=self.coeff[:]
        c2=b.coeff[:]
        coeffs=[]
        for i in range(len(c1)):
            for j in range(len(c2)):
                coeffs.append(c1[i]*c2[j])
        power=[]
        for i in range(len(c1)):
            for j in range(len(c2)):
                power.append((len(c1)-i-1)+(len(c2)-j-1))
        coef_=[]
        for n in range(max(power)+1):
            coef_.insert(0,sum(list( coeffs[i] for i in [j for j in range(len(power)) if power[j] == n] )))
        return polynomial(coef_,self.v)
## Q4
    def plot(self, number_of_points=100, left=-1, right=1):
            #import matplotlib as mpl
            #import matplotlib.pyplot as plt
        increament=(right-left)/number_of_points
        x_space=[]
        for i in range(number_of_points+1):
            x_space.append(left+i*increament)
        coeffs=self.coeff
        y_space=[]
        for i in x_space:
            y_j=[]
            for j in range(len(coeffs)):
                power=len(coeffs)-j-1
                y_j.append(coeffs[j]*i**power)
            y_space.append(sum(y_j))
        plt.plot(x_space,y_space)
        plt.show()
```
